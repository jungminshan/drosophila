```python
matplotlib.rcParams.update({'font.size':20})
matplotlib.rcParams['legend.frameon'] = False

with open('../pickles/LAD_(99,99)_ttv.pkl', 'rb') as f:
    LAD = pickle.load(f)

with open('../pickles/skl_(99,99)_tv.pkl', 'rb') as f:
    skl = pickle.load(f)
    
plt.figure(figsize=(12,6))
plt.plot(range(len(LAD['ferror'])), LAD['ferror'], 'o-', linewidth=2, label='erf-weighted LAD')
# plt.plot(range(1,len(skl['ferror'])+1), skl['ferror'], 'o-', label='LS')
plt.xlim(-1,99)
plt.xticks([0,98],['0','98'])
plt.ylim(0,0.32)
plt.yticks([0,0.1,0.2,0.3],['0','','','0.3'])
plt.legend()
plt.xlabel('Gene')
plt.ylabel('Fractional error')
plt.savefig('LAD_ferror.pdf', bbox_inches='tight')
plt.show()

plt.figure(figsize=(6,5))
error_lad = np.sum(LAD['error'], axis=1)
sns.distplot(error_lad, bins=100, label='erf-weighted LAD')
error_skl = np.sum(skl['error'], axis=1)
ax1=sns.distplot(error_skl, bins=100, label='LS')
plt.xlim(0,12)
plt.xticks([0,3,6,9,12], ['0','','','','12'])
plt.ylim(0,0.7)
plt.yticks([0,0.2,0.4,0.6],['0','','','0.6'])
plt.legend(bbox_to_anchor=(0.3,1))
plt.xlabel('Error')
plt.ylabel('Prbability density')
plt.text(0.5, 1.1, 'On LAD criterion: $L_{1}$ norm', 
         horizontalalignment='center', fontsize=24, transform=ax1.transAxes)

print(stats.mannwhitneyu(error_lad, error_skl, alternative='less'))
# print(np.median(error_lad), np.median(error_skl))

plt.savefig('(99,99)_to_99_L1.pdf', bbox_inches='tight')
# plt.show()

plt.figure(figsize=(6,5))
error_lad = np.sum(LAD['error']**2, axis=1)
sns.distplot(error_lad, bins=100, label='erf-weighted LAD')
error_skl = np.sum(skl['error']**2, axis=1)
sns.distplot(error_skl, bins=100, label='LS')
plt.xlim(-0.1, 3)
plt.xticks([0,1,2,3],['0','','','3'])
plt.ylim(0,4)
plt.yticks([0,2,4],['0','','4'])
plt.legend(bbox_to_anchor=(1,1))
plt.xlabel('Error')
plt.ylabel('Prbability density')
plt.text(0.5, 1.1, 'On LS criterion: $L_{2}^{2}$ norm',
         horizontalalignment='center', fontsize=24, transform=ax1.transAxes)

print(stats.mannwhitneyu(error_lad, error_skl, alternative='less'))
print(np.median(error_lad), np.median(error_skl))

plt.savefig('(99,99)_to_99_L2.pdf', bbox_inches='tight')
plt.show()
```
