```python
from sklearn.model_selection import (TimeSeriesSplit, KFold, ShuffleSplit,
                                     StratifiedKFold, GroupShuffleSplit,
                                     GroupKFold, StratifiedShuffleSplit)
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

matplotlib.rcParams.update({'font.size':20})
matplotlib.rcParams['legend.frameon'] = False

np.random.seed(1338)
cmap_data = plt.cm.Paired
cmap_cv = plt.cm.coolwarm
n_splits = 10

n_points = 100
X = np.random.randn(100, 10)
percentiles_classes = [.1, .3, .6]
y = np.hstack([[ii] * int(100 * perc)
               for ii, perc in enumerate(percentiles_classes)])
groups = np.hstack([[ii] * 10 for ii in range(10)])
def plot_cv_indices(cv, X, y, group, ax, n_splits, lw=15):
    """Create a sample plot for indices of a cross-validation object."""

    # Generate the training/testing visualizations for each CV split
    for ii, (tr, tt) in enumerate(cv.split(X=X, y=y, groups=group)):
        # Fill in indices with the training/test groups
        indices = np.array([np.nan] * len(X))
        indices[tt] = 1
        indices[tr] = 0

        # Visualize the results
        ax.scatter(range(len(indices)), [ii + .5] * len(indices),
                   c=indices, marker='_', lw=lw, cmap=cmap_cv,
                   vmin=-.2, vmax=1.2)

    # Formatting
    yticklabels = list(range(n_splits))
    xticklabels=['0','','','','','6077']
    ax.set(yticks=np.arange(n_splits) + .5, yticklabels=yticklabels, xticklabels=xticklabels,
           xlabel='Cell index', ylabel="CV iteration",
           ylim=[n_splits, -.2], xlim=[0, 100])
    ax.set_title('{}'.format(type(cv).__name__), fontsize=20)
    ax.legend([Patch(color=cmap_cv(0.8)), Patch(color=cmap_cv(.2))],
             ['Testing cells', 'Training cells'], loc=(1.02, 0.8))
    return ax

fig, ax = plt.subplots(figsize=(8,6))
cv = KFold(n_splits)
plot_cv_indices(cv, X, y, groups, ax, n_splits)
plt.savefig('KFold.pdf', bbox_inches='tight')
plt.show()
```

```python
from sklearn.model_selection import (TimeSeriesSplit, KFold, ShuffleSplit,
                                     StratifiedKFold, GroupShuffleSplit,
                                     GroupKFold, StratifiedShuffleSplit)
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

matplotlib.rcParams.update({'font.size':20})
matplotlib.rcParams['legend.frameon'] = False

np.random.seed(1338)
cmap_data = plt.cm.Paired
cmap_cv = plt.cm.coolwarm
n_splits = 10

# n_points = 100
X = np.random.randn(90, 10)
percentiles_classes = [.1, .3, .6]
y = np.hstack([[ii] * int(90 * perc)
               for ii, perc in enumerate(percentiles_classes)])
groups = np.hstack([[ii] * 10 for ii in range(9)])
def plot_cv_indices(cv, X, y, group, ax, n_splits, lw=15):
    """Create a sample plot for indices of a cross-validation object."""

    # Generate the training/testing visualizations for each CV split
    for ii, (tr, tt) in enumerate(cv.split(X=X, y=y, groups=group)):
        # Fill in indices with the training/test groups
        indices = np.array([np.nan] * len(X))
        indices[tt] = 1
        indices[tr] = 0

        # Visualize the results
        ax.scatter(range(len(indices)), [ii + .5] * len(indices),
                   c=indices, marker='_', lw=lw, cmap=cmap_cv,
                   vmin=-.2, vmax=1.2)
        ax.scatter(range(len(X),100), [ii + .5] * 10, 
                   c='k', marker='_', lw=lw, cmap=cmap_cv, vmin=-0.2, vmax=1.2)

    # Formatting
    yticklabels = list(range(n_splits))
    xticklabels=['0','','','','','6077']
    ax.set(yticks=np.arange(n_splits) + .5, yticklabels=yticklabels, xticklabels=xticklabels,
           xlabel='Cell index', ylabel="CV iteration",
           ylim=[n_splits, -.2], xlim=[0, 100])
    ax.set_title('KFold with Validation'.format(type(cv).__name__), fontsize=20)
    ax.legend([Patch(color=cmap_cv(0.8)), Patch(color=cmap_cv(.2)), Patch(color='k')],
             ['Testing cells', 'Training cells', 'Validating cells'], loc=(1.02, 0.8))
    return ax

fig, ax = plt.subplots(figsize=(8,6))
cv = KFold(n_splits)
plot_cv_indices(cv, X, y, groups, ax, n_splits)
plt.savefig('KFold_val.pdf', bbox_inches='tight')
plt.show()
```
