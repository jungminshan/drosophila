```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import matplotlib
```

```python
matplotlib.rcParams.update({'font.size':22})

x = np.linspace(0,10,100)
y = 1.5*x+np.random.normal(0,0.5,100)

# regr = linregress(x,y)

plt.plot(x,y, '.', markersize=12, alpha=0.6)

# x_lin = np.linspace(0,10,2)
# y_lin = x_lin*regr[0] + regr[1]
# plt.plot(x_lin, y_lin, 'r--', linewidth=4)
plt.xlabel('x')
plt.ylabel('y', rotation=0)
plt.xticks([0,5,10],['0','','10'])
plt.yticks([0,5,10,15],['0','','','15'])
plt.text(0.2,13,'y = 1.5x + $\eta$', fontsize=26)
plt.savefig('./paper/out1.pdf', bbox_inches='tight')
plt.show()
```

```python
matplotlib.rcParams.update({'font.size':22})

regr = linregress(x,y)

plt.plot(x,y, '.', markersize=12, alpha=0.6)

x_lin = np.linspace(0,10,2)
y_lin = x_lin*regr[0] + regr[1]
plt.plot(x_lin, y_lin, 'r--', linewidth=4)
plt.xlabel('x')
plt.ylabel('y', rotation=0)
plt.xticks([0,5,10],['0','','10'])
plt.yticks([0,5,10,15],['0','','','15'])
plt.text(0.2,13,'slope = %s' %round(regr[0],2), fontsize=26)
plt.savefig('./paper/out2.pdf', bbox_inches='tight')
plt.show()

print(regr[0], regr[1])
```

```python
xr = np.array([1,5,10])
yr = np.array([10,0,1])
x=np.concatenate((x,xr))
y=np.concatenate((y,yr))

# regr = linregress(x,y)
n = len(xr)
plt.plot(x[:-n],y[:-n], '.', markersize=12, alpha=0.6)
plt.plot(x[-n:], y[-n:], 'r.', markersize=15)

# x_lin = np.linspace(0,10,2)
# y_lin = x_lin*regr[0] + regr[1]
# plt.plot(x_lin, y_lin, 'r--', linewidth=4)
plt.xlabel('x')
plt.ylabel('y', rotation=0)
plt.xticks([0,5,10],['0','','10'])
plt.yticks([0,5,10,15],['0','','','15'])
# plt.text(0.5,12,'y = 1.5x + $\eta$', fontsize=26)
plt.savefig('./paper/out3.pdf', bbox_inches='tight')
plt.show()
```
